```python
import numpy as np
from prettytable import PrettyTable

# configuration variables
# -----------------------------------------------------------
# initial values
a = [1, 0, 0]

# margin vector
b = [1, 2.5, 0.5, 2.5, 2.5, 1]
# dataset
Y = [[1, 0, 2], [1, 1, 2], [1, 2, 1], [-1, 3, -1], [-1, 2, 1], [-1, 3, 2]]

# learning rate
n = 0.1

# iterations
iterations = 12

# dataset
# -----------------------------------------------------------
X = [[0, 2], [1, 2], [2, 1], [-3, 1], [-2, 1], [-3, 2]]
Y = [[1, 0, 2], [1, 1, 2], [1, 2, 1], [-1, 3, -1], [-1, 2, 1], [-1, 3, 2]]

# sequential widrow-hoff learning algorithm
# -----------------------------------------------------------
result = []

print('Length of y is', range(len(Y)))

for o in range(int(iterations / len(Y))):
    print('o is ', o)
    for i in range(len(Y)):
        print('i is ', i)
        a_prev = a
        y = Y[i] #y is 3

        # calculate ay
        ay = np.dot(a, y)
        print('ay is ', ay)

        # calculate update part, below initializes a vector 
        update = np.zeros(len(y))
        
        print('length of update is ', len(update))

        for j in range(len(y)):  #This is the vector update that needs to be added to a 
            update[j] = n * (b[i] - ay) * y[j]
        
        print('Final Update is ', update)
        # add update part to a
        a = np.add(a, update)

        # append result
        result.append((str(i + 1 + (len(Y) * o)), np.round(a_prev, 4), np.round(y, 4), np.round(ay, 4), np.round(a, 4)))

# prettytable
# -----------------------------------------------------------

pt = PrettyTable(('iteration', 'a', 'y', 'ay', 'a_new'))
for row in result: pt.add_row(row)

pt.align['iteration'] = 'c'
pt.align['a'] = 'l'
pt.align['y'] = 'l'
pt.align['ay'] = 'l'
pt.align['a_new'] = 'l'

print(pt)
```
